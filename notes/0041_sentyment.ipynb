```python
import json
from transformers import pipeline
from utils import evaluate_offer_description

# Inicjalizacja modelu klasyfikacji zero-shot
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Ścieżki plików
input_file = "../data/all_offers.json"
output_file = "../data/all_offers_with_scores.json"

# Odczytanie danych z pliku JSON
with open(input_file, "r", encoding="utf-8") as file:
    offers = json.load(file)

# Przetwarzanie każdej oferty
total_offers = len(offers)
for index, offer in enumerate(offers, start=1):
    description = offer.get("opis", [])
    if not description or not description[0]:
        print(f"Pominięto ofertę {index}/{total_offers}: brak opisu.")
        continue
    offer["evaluations"] = evaluate_offer_description(classifier, description[0])
    print(f"Przetworzono ofertę {index}/{total_offers}")

# Zapisanie wyników do nowego pliku JSON
with open(output_file, "w", encoding="utf-8") as file:
    json.dump(offers, file, ensure_ascii=False, indent=4)

print(f"Przetwarzanie zakończone. Wyniki zapisano w {output_file}.")
```
